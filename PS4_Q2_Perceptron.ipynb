```python
import numpy as np
```

```python
values = [
    [1, 4, 1],
    [1, 2, 3],
    [0, 0, 1],
    [-1, 4, 0],
    [1, 0, -2],
    [-1, -1, 1],
    [0, -4, 0],
    [1, 0, -3]
]

y = np.array([1,1,1,1,0,0,0,0])

X = np.array(values)
```

```python
alpha = 0.5 # Learning rate
epochs = 2
w = np.zeros(3)
b = 1
```

```python
for i in range(epochs):
  print("Epoch: ", i+1)
  err = 0
  for j in range(X.shape[0]):
    z = w.T @ X[j] + b
    o = 1 if z>0 else 0
    print(f"z = {z}, o = {o}, y[j] = {y[j]}")
    if y[j] != o:
      w = w + alpha * (y[j] - o) * X[j]
      b = b + alpha * (y[j] - o)
      err += 1
      print("Weights: ", w)
      print("Bias: ", b)

  print("Error: ", err)
  if err == 0:
    break
```
